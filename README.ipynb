```python
#@title Spack: get dependencies
def Deps():
    spec = !"$APPDIR/spack/bin/spack" spec \
        --fresh \
        "$SPEC" \
    ##

    # find "^$package@" names
    deps = []
    for line in spec.l:
        match = auto.re.search(r'\^([^@]+)@', line)
        if match:
            dep = match.group(1)
            # remove ansi color escape codes
            dep = auto.re.sub(r'\x1b\[[0-9;]*m', '', dep)
            print(dep)
            deps.append(dep)

    return deps

deps = Deps()
```

## Spack: Load

```python
%%scope
#@title Spack: load
env = {}
# !"$APPDIR/spack/bin/spack" load --sh "adios2+python~bzip2" \
#     >"$tmp" \
# ##

stdout = auto.subprocess.check_output(
    [
        f"{APPDIR}/spack/bin/spack", "load",
        "--sh",
        f"{SPEC}",
    ],
    stderr=auto.subprocess.STDOUT,
    text=True,
)

lines = stdout.splitlines()
for line in lines:
    line = line.strip()
    if line:
        # print(line)
        args = auto.shlex.split(line)
        if args[0] == 'export':
            key, value = args[1].split('=')
            env[key] = value

auto.pprint.pprint(env)
auto.os.environ |= env
```
